```r
df <- read_excel("Tomate.xlsx")

colnames(df) <- c("estacao", "tratamento", "bloco", "colheita", "peso", "n_frutos", "comprimento", "largura", "pca")

head(df, 3)
```

```r
bioproduto <- ifelse(
    df$tratamento <= 2,
    0.5,
    ifelse(
        df$tratamento <= 4,
        1,
        ifelse(
            df$tratamento <= 6,
            1.5,
            ifelse(
                df$tratamento <= 8,
                2,
                ifelse(
                    df$tratamento <= 10,
                    2.5,
                    ifelse(
                        df$tratamento == 11,
                        5,
                        10
                    )
                )
            )
        )
    )
)

alternada <- ifelse(
    df$tratamento %in% c(1,3,5,7,9,11,12),
    0,
    1
)

df$bioproduto <- bioproduto
df$alternada <- alternada

df <- df %>% filter(estacao == 'P-V')
df <- df %>% select(c(tratamento, bloco, colheita, n_frutos, peso, comprimento, largura, bioproduto, alternada))
df <- df %>% arrange(tratamento, bloco, colheita)
```

```r
head(df, 3)
```

```r
df_aux <- df %>%
group_by(tratamento, bloco) %>%
summarise(  
    colheita = list(colheita),
    n_frutos = list(n_frutos),
    peso = list(peso),
    comprimento = list(comprimento),
    largura = list(largura)
)
head(df_aux,3)
```

```r
n_frutos_cum <- lapply(df_aux$n_frutos, function(list_n_frutos){
    cumsum(unlist(list_n_frutos))
})

df_aux$n_frutos_cum <- n_frutos_cum
head(df_aux, 3)
```

```r
df <- (
    df_aux %>%
    unnest(
        c(colheita, n_frutos, n_frutos_cum, peso, comprimento, largura),
        names_repair = "unique"
    ) %>%
    select(c(tratamento, bloco, colheita, n_frutos, n_frutos_cum, peso, comprimento, largura))
)
```

```r
write.csv(df, "df_n_frutos_cum.csv", row.names = FALSE)
```

```r
df_aux <- read.csv("df_aux.csv")

head(df_aux,6)
```

```r
options(repr.plot.width=16, repr.plot.height=8)
ggplot(df_aux)+
    geom_boxplot(aes(x = tratamento, y = n_frutos_cum, color = as.factor(bloco)))+
    geom_point(aes(x = tratamento, y = n_frutos_cum, color = as.factor(bloco) ))
```

```r
df_aux$tratamento <- factor(df_aux$tratamento)
df_aux$bloco <- factor(df_aux$bloco)

fit <- lm(n_frutos_cum ~ tratamento + bloco, data = df_aux)
summary(fit)
```

```r
model <- aov(n_frutos_cum ~ tratamento + bloco, data = df_aux)
summary(model)
```
